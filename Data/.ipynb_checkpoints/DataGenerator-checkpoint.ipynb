```python
num_MA = 3

raw_MA_columns = [f'{num_MA} MA {i}' for i in range(1, 31)]
raw_financial_columns = ['rgs_dt', 'tck_iem_cd', 'label', 'score', 'cluster', 'industry']
raw_df = df.loc[:, df.columns[df.columns.str.startswith(tuple(raw_MA_columns)) | df.columns.isin(raw_financial_columns)]]
raw_df.industry = raw_df.industry.astype(int)
raw_df = raw_df.drop_duplicates()
```

```python
def get_stock_info(tck, event_date):
    start_dt = event_date + pd.DateOffset(days=3)  # 3일 이동평균이라고 가정
    end_dt = start_dt + pd.DateOffset(days=60)
    
    info = fdr.DataReader(tck,start_dt, end_dt).drop(['Close'],axis=1)
    
    return info.head(30)

def get_MA_prediction_info(df, tck, event, sent, industry):
    # df = 'MA_predict_3days.csv'
    predict_info = df[ (df.event==event)&(df.sent==sent)&(df.industry==industry)]
    ma_prediction = predict_info.drop(['event','sent','industry'],axis=1).reset_index(drop=True).T
    
    return ma_prediction
    
    
    
    
    
    
    
```

```python
# # def fed_fund_rate(event_date):
#     start_dt = event_date + pd.DateOffset(days=3)  # 3일 이동평균이라고 가정
#     end_dt = start_dt + pd.DateOffset(days=30)
# #     rf = pdr.get_data_fred('FEDFUNDS', start_dt, end_dt)
# #     return rf


def currency_rate_EUR(event_date):

    start_dt = event_date + pd.DateOffset(days=3)  # 3일 이동평균이라고 가정
    end_dt = start_dt + pd.DateOffset(days=60)
    
#     # 주말 및 공휴일을 제외한 개장일을 가져옴
#     schedule = nyse.valid_days(start_date=start_dt, end_date=end_dt)
    
    # 개장일에 대한 데이터를 불러옴
    EUR = pd.DataFrame(
        fdr.DataReader('EUR/USD', start=start_dt, end=end_dt)[['Close']].rename(columns={"Close": "EUR/USD"}))
    
    
    return EUR.head(30)
```

```python
def save_file(result,industry, dt, tck):
    
    industry_folder = os.path.join(os.getcwd(), f'Industry_{industry}')
    os.makedirs(industry_folder, exist_ok=True)

    # Create a folder for the ticker inside the industry folder (if not exists)
    ticker_folder = os.path.join(industry_folder, f'{tck}')
    os.makedirs(ticker_folder, exist_ok=True)

    # Save the result as a CSV file in the ticker folder
    csv_file_path = os.path.join(ticker_folder, f'{dt}.csv')
    result.to_csv(csv_file_path, index=False)

    print(f"Result saved for Industry {industry}, Ticker {tck} on {dt} to {csv_file_path}")
```

```python


def divide_by_event(tck_df, tck):
    ma_data = pd.read_csv('MA_predict_3days.csv').reset_index(drop=True)
    ma_data.drop(['Unnamed: 0'],axis=1, inplace=True)
    # 기존 파일 => 이동평균 시계열 데이터로 변환
    start_col = 5
    end_col = 35

    ma_col = tck_df.iloc[:, list(range(1)) + list(range(5, 35))]
    news_col = tck_df.loc[:, ['rgs_dt', 'tck_iem_cd', 'label', 'score', 'industry', 'cluster']]

    tm = ma_col.T.rename(columns=ma_col.T.iloc[0])
    tm.drop(['rgs_dt'], axis=0, inplace=True)
    tm = tm.loc[:, ~tm.columns.duplicated()]

    tn = news_col.T.rename(columns=news_col.T.iloc[0])
    tn.drop(['rgs_dt'], axis=0, inplace=True)
    tn = tn.loc[:, ~tn.columns.duplicated()]

    events_byTicker = tn.columns.tolist()
    
    for dt in events_byTicker:
        dt = pd.to_datetime(dt)
        tmp = tm.loc[:, dt]
        label = tn.loc['label', dt]
        score = tn.loc['score', dt]
        industry = tn.loc['industry', dt]
        cluster = tn.loc['cluster', dt]

        result = pd.DataFrame({
            'label': label,
            'score': score,
            'industry': industry,
            'event': cluster,
            '3 MA': tmp
        })

        tmp = tmp.reset_index(drop=True)
        
        result = result.reset_index(drop=True)
        currency_rate = currency_rate_EUR(dt).reset_index(drop=True)
        stock_info = get_stock_info(tck,dt).reset_index()
        ma_prediction = get_MA_prediction_info(df=ma_data, tck=tck, event=cluster, sent=label, industry=industry).reset_index(drop=True)

        
        ma_ratio = pd.DataFrame({'MA_ratio':(tmp/ma_prediction)[0]})
        
        
        result = pd.concat([stock_info,result,ma_ratio, currency_rate],axis=1)
        
        save_file(result,industry,dt, tck)

    return result
```

```python
if __name__ == '__main__':
    
    for tck in tck_list:
        tck_df = raw_df[ raw_df.tck_iem_cd == tck]
        tck_df = tck_df.drop_duplicates()
        divide_by_event(tck_df, tck)
```
